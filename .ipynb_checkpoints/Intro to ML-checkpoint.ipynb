```python
import scipy as sp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

fruits = pd.read_table('fruit_data_with_colors.txt')
```

```python
fruits.head()
```

```python
look_up_fruit_name = dict(zip(fruits.fruit_label.unique(), fruits.fruit_name.unique()))
look_up_fruit_name
```

```python
# creating test set, split train set
X = fruits[['mass','width','height','color_score']]
y = fruits['fruit_label']

X_train, X_test , y_train, y_test = train_test_split(X,y, random_state=80)
```

```python
y_test.head()
```

```python
from matplotlib import cm
cmap = cm.get_cmap('gnuplot')
scatter = pd.scatter_matrix(X_train, c=y_train, marker='*', s=55, hist_kwds={'bins':20}, figsize=(8,8), cmap=cmap)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter(X_train['width'], X_train['height'], X_train['color_score'], c = y_train, marker='.', s=100)
ax.set_xlabel('width')
ax.set_ylabel('height')
ax.set_zlabel('color')
plt.show()
```

```python
### implementing K-neighbor classifier

from sklearn.neighbors import KNeighborsClassifier

#setting a ceratin value to b the classifier
knn= KNeighborsClassifier(n_neighbors=6)
```

```python
# trainig the classifier using the training set
knn.fit(X_train, y_train)
```

```python
#estimating the accuracy of the classifier using the test data
knn.score(X_test, y_test)
```

```python
### using logistic regression
from sklearn import linear_model
logistic = linear_model.LogisticRegression(C=1)
logistic.fit(X_train, y_train)
```

```python
#estimating the accuracy of the classifier using the test data
logistic.score(X_test, y_test)
```

```python
# using the new classifier to check classification of new fruits using Knn classifier

fruit_prediction = knn.predict([[80,4.5,5.5,0.70]])
lookup_fruit_name=[fruit_prediction[0]]
print('the label of the predicted fruit',lookup_fruit_name)
name = fruits.loc[fruits['fruit_label']==lookup_fruit_name].fruit_name.values[0]
print('Classified as',name)
```

```python
# using the new classifier to check classification of new fruits using knn classier

fruit_prediction = knn.predict([[80,4.5,5.5,0.70]])
lookup_fruit_name=[fruit_prediction[0]]
print('the label of the predicted fruit',lookup_fruit_name)
name = fruits.loc[fruits['fruit_label']==lookup_fruit_name].fruit_name.values[0]
print('Classified as',name)
```

```python
# using the new classifier to check classification of new fruits using LOGISTIC regression

fruit_prediction = logistic.predict([[80,4.5,5.5,0.70]])
lookup_fruit_name=[fruit_prediction[0]]
print('the label of the predicted fruit',lookup_fruit_name)
name = fruits.loc[fruits['fruit_label']==lookup_fruit_name].fruit_name.values[0]
print('Classified as',name)
```
